## Beta

```python
Returns= pd.read_excel('./data/行业收益率.xlsx',skiprows=[1,2],index_col="日期") 
Returns
```

```python
Returns=Returns.applymap(lambda x:x/100)
Returns 
```

```python
HSI=pd.read_excel('./data/HSI.xlsx',index_col="Date")
HSI
```

```python
hebing=HSI.merge(Returns,left_index=True,right_index=True).dropna(axis=1,how='all')
hebing
```

```python
hangye_rf=hebing-rf
hangye_rf.head()
```

```python
import statsmodels.api as sm
def regress(data, yvar, xvars):     
    data=pd.concat([data[yvar], data[xvars]], axis=1)
    Y = data[yvar]     
    x = data[xvars] 
    X = sm.add_constant(x)
    result = sm.OLS(Y, X,missing='drop').fit()     
    return yvar,result.params[1]
#     print('{}的beta为：{}'.format(yvar,result.params[1]))
```

```python
beta=[]
for i in hangye_rf.columns[1:]:
    j=regress(hangye_rf,i,'HSI')
    beta.append(j)
beta
```

```python
Betas=pd.Series(dict(beta))
Betas 
```

```python
Betas.apply(lambda x:'{:.2}'.format(x))
```

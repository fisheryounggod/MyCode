## 性能测试

````python
import time 
import pandas as pd 
import modin.pandas as mpd 
````

````python
%time df_pd=pd.read_csv('./data/Earthquakes.csv')
````

````python
%time df_mpd=mpd.read_csv('./data/Earthquakes.csv')
````

````python
%time pd.concat([df_pd for _ in range(5)], ignore_index=True)
````

````python
%time mpd.concat([df_mpd for _ in range(5)], ignore_index=True)
````

````python
%%time
df_pd.isna().any()
````

````python
%%time
df_mpd.isna().any()
````

modin在处理大型数据集时，部分应用场景可以用其替换pandas

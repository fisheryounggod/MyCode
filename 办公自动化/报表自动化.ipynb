这个表要包括销售额、客流量、客单价这三个指标的本月累计、上月同期、去年同期、环比、同比这几个数值。

```python
# 指标说明

- 销售额=单价$\times$销量
- 客流量=订单ID去重计数 
- 客单价=销售额/客流量
- 本月=2018 年 2 月
- 上月=2018 年 1 月
- 去年同期=2017 年 2 月
```

```python
import pandas as pd 
import numpy as np
from datetime import datetime 
from dateutil.parser import parse
```

```python
df = pd.read_csv("./Excel-Python-DA/order-14.1.csv", encoding='gbk', parse_dates=['成交时间']).dropna(how='all')
df #parse_dates参数表示将数据解析为时间格式。
```

```python
df.info()
```

## 傻算

### 计算本月相关指标

```python
df_this_month = df[(df['成交时间']>=datetime(2018,2,1))&(df['成交时间']<=datetime(2018,2,28))]
df_this_month
```

```python
本月销售额 = (df_this_month['单价']*df_this_month['销量']).sum()
本月销售额
```

```python
客流量 = df_this_month['订单ID'].drop_duplicates().count()
客流量
```

```python
客单价 = 本月销售额/客流量
客单价.round(2)
```

### 计算上月相关指标

```python
last_month = df[(df['成交时间']>=datetime(2018,1,1))&(df['成交时间']<=datetime(2018,1,31))]
last_month

上月销售额 = (last_month['单价']*last_month['销量']).sum()
上月销售额

上月客流量 = last_month['订单ID'].drop_duplicates().count()
上月客流量

上月客单价 = 上月销售额/上月客流量
上月客单价.round(2)
```

## 函数式编程

### 按需提取并计算

```python
def get_df(df, y, m):
    df['年'] = df['成交时间'].apply(lambda x: x.strftime('%Y'))
    df['月'] = df['成交时间'].apply(lambda x: x.strftime('%m'))
    need_month = df[(df['年'] == str(y)) & (df['月'] == str(m))]
    return need_month

get_df(df,2018,'02')
```

### 统一处理并生成报告

```python
def jisuan(df):
    xiaoshoue = (df['单价']*df['销量']).sum()
    traffic = df['订单ID'].drop_duplicates().count()
    danjia = xiaoshoue/traffic
    return xiaoshoue,traffic,danjia
```

```python
sale_1,traffic_1,s_t_1 = jisuan(get_df(df,2018,'02'))

sale_2,traffic_2,s_t_2 = jisuan(get_df(df,2018,'01'))

sale_3,traffic_3,s_t_3 = jisuan(get_df(df,2017,'02'))
```

```python
report = pd.DataFrame([[sale_1,sale_2,sale_3],
                       [traffic_1,traffic_2,traffic_3],
                       [s_t_1,s_t_2,s_t_3]],
                      columns=['本月累计','上月同期','去年同期'],
                      index=['销售额','客流量','客单价'])
report
```

### 添加环比

```python
report['环比']= report['本月累计']/report['上月同期']-1
report
```

### 添加同比

```python
report['同比']= report['本月累计']/report['去年同期']-1
report
```

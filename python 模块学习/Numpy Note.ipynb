## 数据清洗

```python
def get_repeated_label(array_1dim):   # 查找重复
    label_1dim = []
    for k in range(len(array_1dim) - 1):
        for i in range(k + 1, len(array_1dim), 1):
            if array_1dim[k] == array_1dim[i]:
                label_1dim.append(i)
    return np.unique(label_1dim)
```

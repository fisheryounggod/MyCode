**基本原理**：想生成几簇数据，就创建几条线段，然后在y上做一个高斯偏移。

```python
plt.figure(figsize=(12,8))
blob1 = 500
x1 = np.linspace(0.5,3,blob1)
y1 = 2 * x1+10+stats.norm.rvs(0.,2.,size=(blob1,))

blob2 = 800
x2 = np.linspace(5,8,blob2)
y2 = 2*x2 -4+stats.norm.rvs(0.,3.0,size=(blob2,))

blob3 = 100
x3 = np.linspace(3,4,blob3)
y3 = 2*x3 -4+stats.norm.rvs(0.,1.0,size=(blob3,))

plt.scatter(x1,y1,label='cluster0')
plt.scatter(x2,y2,label='cluster1')
plt.scatter(x3,y3,label='cluster2')
plt.legend()
plt.show()
```

```python
myDF = stats.norm(5,3) # Create the frozen distribution
```

```python
x = np.linspace(-5, 15, 101) 
fig=plt.figure()
ax1=fig.add_subplot(2,2,1)
ax1.plot(x,myDF.cdf(x)) # Calculate the corresponding CDF
ax2=fig.add_subplot(2,2,2)
ax2=plt.plot(x,myDF.ppf(x))
ax3=fig.add_subplot(2,2,3)
ax3.plot(x,myDF.sf(x))
ax4=fig.add_subplot(2,2,4)
ax4.plot(x,myDF.isf(x))
```

```python
y=myDF.cdf(x)
z=myDF.sf(x)
u=myDF.pdf(x)
v=myDF.ppf(x)
w=myDF.isf(x)
```

```python
plt.subplot(2,3,1),plt.plot(x,y),plt.title('CDF')
plt.subplot(2,3,2),plt.plot(x,z),plt.title('SF')
plt.subplot(2,3,3),plt.plot(x,u),plt.title('PDF')
plt.subplot(2,3,4),plt.plot(x,v),plt.title('PPF')
plt.subplot(2,3,5),plt.plot(x,v),plt.title('ISF')
plt.subplots_adjust(wspace=0,hspace=0)
```

### Bernoulli Distribution

```python
bernoulliDist.pmf(0)
```

```python
bernoulliDist.pmf(1)
```

```python
bernoulliDist.rvs(10) #simulate 10 Bernoulli trials
```

```python
(p, num) = (0.5, 4)
```

```python
myDistribution.ppf( [significanceLevel/2, 1-significanceLevel/2] )
```

## 快速傅里叶变换：scipy.fftpack

```python
time_step = 0.02
```

```python
period = 5. 
```

```python
time_vec = np.arange(0, 20, time_step)
```

```python
sig = np.sin(2 * np.pi / period * time_vec) + 0.5 * np.random.randn(time_vec.size)
```

```python
time_vec
```

```python
sig 
```

```python
from scipy import fftpack
```

```python
sample_freq = fftpack.fftfreq(sig.size, d=time_step)
```

```python
sig_fft = fftpack.fft(sig)
sig_fft 
```

```python
sample_freq
```

## 优化器：scipy.optimize

```python
def f(x):
    return x**2+10*np.sin(x)
```

```python
x = np.arange(-10, 10, 0.1)
```

```python
plt.plot(x, f(x))
```

```python
from scipy import optimize 
optimize.fmin_bfgs(f,0)
```

```python
optimize.fmin_bfgs(f,3,disp=0)
```

```python
import scipy
scipy.optimize.basinhopping(f,3,disp=0)
```

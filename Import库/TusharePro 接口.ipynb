## 获取A股基本信息

```python
def basic():
    """获取当前上市的股票代码、简称、注册地、行业、上市时间等数据"""
    import tushare as ts
    token='693c84c2aca4dc7b941439006bd946413706e7a9560bf3cce0369bb3'
    #ts.set_token(token)
    pro = ts.pro_api(token)
    return pro.stock_basic(list_status='L')
```

## 根据股票名称查股票代码

```python
def findcode(name):
    """根据股票名称查股票代码"""
    basic=basic()
    return basic[basic.name==name].ts_code.tolist()[0]
```

## Chapetr 2 金融时间序列的线性模型

- 严平稳：$(x_1,x_2,\dots, x_{k})$和$(x_{1+h},x_{2+h},\dots, x_{k+h})$联合分布相同，即联合分布只与变量间隔有关。

- 弱平稳:一个时间序列$x_t$的一阶矩和二阶矩具有时间不变性。

    i）$E(x_t^2) < \inf$ (二阶矩存在)

    ii）$E(x_t)=\mu$ (期望不随时间变化)

    iii) $cov(x_t,x_{t+h})=\sigma _h^2$ （协方差不时变，仅依赖间隔$h$）

### 相关系数

$$\rho_{x,y}=\frac{Cov(X,Y)}{\sqrt{Var(X)Var(Y)}}=\frac{E[(X-\mu_x)(Y-\mu_y)]}{\sqrt{E(X-\mu_x)^2E(Y-\mu_xy)^2}}$$

### 自相关函数（ACF）

- 自相关系数：$$\rho_k=\frac{Cov(x_t,x_{t-k})}{\sqrt{Var(x_t)Var(x_{t-k})}}=\frac{Cov(x_t,x_{t-k})}{Var(x_t)}=\frac{\gamma_k}{\gamma_0}$$

- 自相关函数：自相关系数组成的集合$\{\rho_k\}$

- $x_t$的间隔为k的样本自相关系数：$$\hat{\rho_k}=\frac{\Sigma^T_{t=k+1}(x_t-\bar x)(x_{t-k}-\bar x)}{\sum^T_{t=1}(x_t-\bar x)^2},0\leq k\leq T-1$$

### 白噪声

- 白噪声序列：$x_t$是有限均值、有限方差的独立同分布随机变量变量序列。
- 高斯白噪声：$x_t$还服从均值为零、方差为$\sigma^2$的正态分布。
- 对于白噪声序列，所有的自相关函数为零。

### 线性时间序列

$$x_t=\mu + \sum^\infty_{t=0}\psi_ia_{t-i}$$

> - $a_t$是零均值独立同分布并且有合理定义的分布的随机变量序列，也即$\{a_t\}$是白噪声序列。
> - 系数$\psi_i$决定了$x_t$的动态结构，称为$x_t$的$\psi$权重。
> - 若$x_t$弱平稳，利用$\{a_t\}$的性质，可得$x_t$的均值和方差
$$E(x_t)=\mu, $$
$$Var(x_t)=\sigma_a^2\sum^{\infty}_{i=0}\psi_i^2$$

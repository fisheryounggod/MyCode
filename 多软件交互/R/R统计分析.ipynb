```python
# 

import statsmodels.api as sm

x = df['']

X = sm.add_constant(x)

y = df['']

model = sm.OLS(y,X)

results = model.fit()
```

```python
results.summary()
```

```python
y_fitted = results.fittedvalues

fig, ax = plt.subplots(figsize=(8,6))      # 86 

ax.plot(x, y, 'o', label='data')           # 

ax.plot(x, y_fitted, 'r--.',label='OLS')   # 

ax.legend(loc='best')                       # 
```

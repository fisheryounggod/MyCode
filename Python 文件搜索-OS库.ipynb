## 打造文件搜索工具

```python
pwd=%pwd
```

```python
import os

path = pwd
for file in os.scandir(path): 
    print(file.name, file.path, file.is_dir())
```

**最后输出的是给定路径下各内容的名字、绝对路径，并判断其是否是文件夹**

```python
import os
wenjianjia=[]
num = 0
for file in os.scandir(pwd):
    if file.is_dir():
        wenjianjia.append(file)
    if 'Python' in file.name:
        num += 1
print('{}中文件夹为：{}'.format(pwd,wenjianjia))
print('含有Python的文件夹个数为：', num)
```

```python
import os 
filenum=0
for file in os.listdir(pwd):
    if file.endswith('.ipynb'):
        filenum+=1
print(filenum)
```

```python
# 遍历
import os
for dirpath, dirnames, filenames in os.walk(pwd):
    print(f'打开文件夹{dirpath}')
    if dirnames:
        print(dirnames)
    if filenames:
        print(filenames)
    print('-' * 10)
```

```python
import os
for dirpath, dirnames, filenames in os.walk(pwd):
    if filenames:
        for i in filenames:
            if i.endswith('.xlsx'):
                print(os.path.join(dirpath, i))
```

```python
# glob 遍历支持正则表达式
import glob
for file in glob.glob('**/*.xlsx', recursive=True): 
    print(file)
```

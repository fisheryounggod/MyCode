## 变量的存储类型

*- 整数的存储类型
  * byte     字节型     (-100, +100)
  * int      一般整数型 (-32000, +32000)
  * long     长整数型   (-2.14*10^10, +2.14*10^10)，即，正负21亿 
*- 小数的存储类型
  * float    浮点型   8 位有效数字
  * double   双精度   16位有效数字

## 变量的名称
  * 由英为字母、数字或 _ 组成，至多不超过 32 个；
  * 首字母必须为 字母 或 _ ；
  * 英文字母的大写和小写具有不同的含义；
  * 例如：abc_1 a1 _a2 _Gdp_ 都是合理的变量名
          5gdp 2invest 则不是；
  * ^特别注意^：建议不要使用 _ 作为变量的第一个字母，
  *            因为许多stata的内部变量都是以 _ 开头的，
  *            如，_n, _N, _cons, _b 等等。

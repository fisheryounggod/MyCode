# 第二讲 命令

### 命令格式

```stata
// [by varlist:] command [varlist] [=exp] [if exp] [in range] [weight] [,options]
```

1. [by varlist] ：分类操作

2. command [varlist] ：命令

3. [=exp] ：赋值及运算表达式

4. [if exp] ：条件表达式

5. [in range] ：范围筛选

6. [weight] ：加权
 
7. [,options] ：其他可选项

### 常用命令

```stata
* 帮助查找
help
search


* 打开和保存数据
clear
use
save

* 导入数据
input
edit
infile
insheet

* 重整数据
append
merge
xpose
reshape
generate
egen//?
rename
drop
keep
sort
encode
decode
order
by

* 报告数据
describe
codebook
list
count
inspect
table
tabulate

* 显示和保存结果
display
log
```

## by:分类操作

```stata
clear 
input x y 
1 1.1 
1 1.2 
1 1.3 
2 2.1 
2 2.2 
end
```

```stata
gen n = _n
gen N = _N
gen z = y[1]
l 
```

```stata
by x,sort:gen n1=_n
by x,sort:gen N1=_N
by x,sort:gen z1=y[1]
list 
```

```stata
clear 
input hhid age 
1 86
1 42
1 36
1 57
1 28
2 42 
2 5
2 40
end 
save family, replace 
```

```stata
use family,clear
by hhid,sort:gen hhsize=_N // 得到家庭规模
by hhid,sort:gen id=_n+hhid*10 // 编码
by hhid,sort:egen mage=mean(age) // 平均年龄
sort hhid age //每户内按年龄大小排序
by hhid:gen nid=_n // 户内排序
```

## 伪随机数

```stata
di uniform() //(0,1)均匀分布
```

```stata
di int(10*uniform())
```

```stata
scalar a = 2
scalar b = 3 
di a+(b-a)*uniform() //生成任意均匀分布（a,b）
```

```stata
scalar a = 4
scalar b = 8
di a+int((b-a)*uniform()) //生成任意均匀分布（a,b）整数随机数
```

```stata
set seed 1234
di uniform()
```

```stata
set seed 1234
di uniform()
```
